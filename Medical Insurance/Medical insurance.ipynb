```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/sca-programming-school/datasets/main/insurance.csv')
df.info()
```

```python
df['sex'][df['sex'] == 'female'] = 0
df['sex'][df['sex'] == 'male'] = 1
df['sex'] = df['sex'].astype('int64')
```

```python
df['smoker'][df['smoker'] == 'no'] = 0
df['smoker'][df['smoker'] == 'yes'] = 1
df['smoker'] = df['smoker'].astype('int64')
```

```python
df['region'][df['region'] == 'southwest'] = 0
df['region'][df['region'] == 'southeast'] = 1
df['region'][df['region'] == 'northwest'] = 2
df['region'][df['region'] == 'northeast'] = 3
df['region'] = df['region'].astype('int64')
```

```python
df.info()
```

```python
x = df[['age', 'sex', 'bmi', 'children', 'smoker', 'region']]
x = sm.add_constant(x)
y = df['charges']
model = sm.OLS(y, x).fit()
model.summary()
```

```python
x = df[['age', 'bmi', 'children', 'smoker']]
x = sm.add_constant(x)
y = df['charges']
model = sm.OLS(y, x).fit()
model.summary()
```

```python
test_data = pd.DataFrame({
    'age': [19, 25],
    'bmi': [18.72, 25.57],
    'children': [0, 2],
    'smoker': [0, 0]
})
test_data = sm.add_constant(test_data)
model.predict(test_data)
```
