```python
# Enter your code
df['revenue'] = df['gross'] - df['budget']
# Show the first 5 rows again to check
df.head()
```

```python
# Enter your code
df['percentage_profit'] = df['revenue'] / df['gross'] * 100
```

```python
# Show the first 5 rows again to check
df.head()
```

```python
# Enter your code
df['high_budget_movie'] = df['budget'] > 100000000
```

```python
# Show the first 5 rows again to check
df.head()
```

```python
# Enter your code
df['successful_movie'] = df['percentage_profit'] > 50
# Show the first 5 rows again to check
df.head()
```

```python
# Enter your code
df['high_rated_movie'] = df['score'] > 8
# Show the first 5 rows again to check
df.head()
```

```python
# Enter your code
df['is_new_release'] = df['year'] > 2020
# Show the first 5 rows again to check
df.head()
```

```python
# Enter your code
df['is_long_movie'] = df['runtime'] > 150
# Show the first 5 rows again to check
df.head()
```

```python
# Your code
df['age'] = 2024 - df['year']
# Show the first 5 rows again to check
df.head()
```

```python
# Your code
sorted_df = df.sort_values(by='score', ascending=False)
top10_movies = sorted_df.head(10)
# Your code
top10_movies
```
