```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/sca-programming-school/datasets/main/sales_advertising.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['tv_ads'] = df['tv_ads'].fillna(df['tv_ads'].median())
df['radio_ads'] = df['radio_ads'].fillna(df['radio_ads'].median())
df['newspaper_ads'] = df['newspaper_ads'].fillna(df['newspaper_ads'].median())
df['digital_marketing'] = df['digital_marketing'].fillna(df['digital_marketing'].median())
df = df.dropna(subset=['sales'])
df.info()
```

```python
tv_Q1 = df['tv_ads'].quantile(0.25)
tv_Q3 = df['tv_ads'].quantile(0.75)
tv_IQR = tv_Q3 - tv_Q1
tv_con1 = tv_Q1 - 1.5 * tv_IQR <= df['tv_ads']
tv_con2 = df['tv_ads'] <= tv_Q3 + 1.5 * tv_IQR
df[tv_con1 & tv_con2]
```

```python
radio_Q1 = df['radio_ads'].quantile(0.25)
radio_Q3 = df['radio_ads'].quantile(0.75)
radio_IQR = radio_Q3 - radio_Q1
radio_con1 = radio_Q1 - 1.5 * radio_IQR <= df['radio_ads']
radio_con2 = df['radio_ads'] <= radio_Q3 + 1.5 * radio_IQR
df[radio_con1 & radio_con2]
```

```python
newspaper_Q1 = df['newspaper_ads'].quantile(0.25)
newspaper_Q3 = df['newspaper_ads'].quantile(0.75)
newspaper_IQR = newspaper_Q3 - newspaper_Q1
newspaper_con1 = newspaper_Q1 - 1.5 * newspaper_IQR <= df['newspaper_ads']
newspaper_con2 = df['newspaper_ads'] <= newspaper_Q3 + 1.5 * newspaper_IQR
df[newspaper_con1 & newspaper_con2]
```

```python
digital_Q1 = df['digital_marketing'].quantile(0.25)
digital_Q3 = df['digital_marketing'].quantile(0.75)
digital_IQR = digital_Q3 - digital_Q1
digital_con1 = digital_Q1 - 1.5 * digital_IQR <= df['digital_marketing']
digital_con2 = df['digital_marketing'] <= digital_Q3 + 1.5 * digital_IQR
df[digital_con1 & digital_con2]
```

```python
sales_Q1 = df['sales'].quantile(0.25)
sales_Q3 = df['sales'].quantile(0.75)
sales_IQR = sales_Q3 - sales_Q1
sales_con1 = sales_Q1 - 1.5 * sales_IQR <= df['sales']
sales_con2 = df['sales'] <= sales_Q3 + 1.5 * sales_IQR
df[sales_con1 & sales_con2]
```

```python
df_clean = df[tv_con1 & tv_con2]
x = df_clean[['tv_ads', 'radio_ads', 'newspaper_ads', 'digital_marketing']]
y = df_clean['sales']
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
model.summary()
```

```python
tv_mean = df_clean['tv_ads'].mean()
radio_mean = df_clean['radio_ads'].mean()
digital_mean = df_clean['digital_marketing'].mean()
total = tv_mean + radio_mean + digital_mean
test_data = pd.DataFrame({
    'tv_ads': [tv_mean, total, 0, 0, 0, 0],
    'radio_ads': [radio_mean, 0, total, 0, 0, 0],
    'newspaper_ads': [0, 0, 0, total, 0, 0],
    'digital_marketing': [digital_mean, 0, 0, 0, total, 0]
})
test_data
```

```python
test_data = sm.add_constant(test_data)
model.predict(test_data)
```
